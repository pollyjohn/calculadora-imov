```python
import pandas as pd
```

```python
df = pd.read_csv("raw.csv")
```

```python
df
```

```python
df[df.duplicated(subset=['id','header','address','area','rooms','price'])]
```

```python
df = df.drop_duplicates(subset=['id','header','address','area','rooms','price'], keep = 'first')
```

```python
df.info()
```

```python
filtro_de_anuncio = [w.isnumeric() for w in df['id']]
```

```python
df = df[filtro_de_anuncio]
```

```python
df = df.reset_index(drop=True)
```

```python
df['id'] = df['id'].astype(int)
```

```python
df.info()
```

AREA

```python
#ao fazer o split de letras e numeros que estão como uma só
#string e depois modificando para int, aotomaticamente  é excluido letras 
df['rooms'] = df['rooms'].str.split(' ').str[0].str.replace('--','0').astype(int)
```

```python
df['rooms']
```

```python
df.info()
```

BANHEIROS

```python
df['bathrooms'] = df['bathrooms'].str.split(' ').str[0].str.replace('--','0').astype(int)
```

```python
df['bathrooms']
```

```python
df.info()
```

```python
df['garages'] = df['garages'].str.split(' ').str[0].str.replace('--','0').astype(int)
```

```python
df['garages']
```

preço

```python
df['price'] = df['price'].str.split(' ').str[1].str.replace('.','').astype(int)
```

```python
df.info()
```

```python
df['condo'] = df['condo'].str.split(' ').str[1].str.replace('.','').astype(float) 
```

datetime

```python
df['crawled_at'] =pd.to_datetime(df['crawled_at'], format = '%Y-%m-%d %H:%M:%S')
```

```python
df['crawled_at']
```
